### Cumulative Moving average

```python
df1['CMA'] = df1['average'].expanding(center = False).mean()
df1.head()
```

```python
colors = ['green', 'orange']

# line plot - the yearly average air temperature in Barcelona
df1[['average', 'CMA']].plot(color=colors, linewidth=3, figsize=(12,6))

# modify ticks size
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.legend(labels =['Average', 'CMA'], fontsize=14)

# title and labels
plt.title('IEX', fontsize=20)
plt.xlabel('Entries', fontsize=16)
plt.ylabel('average', fontsize=16)
```

```python
df1['CMA_S'] = df1['Sum'].expanding(center = False).mean()
df1.head()
```

```python
colors = ['green', 'orange']

# line plot - the yearly average air temperature in Barcelona
df1[['Sum', 'CMA_S']].plot(color=colors, linewidth=3, figsize=(12,6))

# modify ticks size
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.legend(labels =['Sum', 'CMA'], fontsize=14)

# title and labels
plt.title('IEX', fontsize=20)
plt.xlabel('Entries', fontsize=16)
plt.ylabel('Sum', fontsize=16)
```

### Exponential Moving average

```python
df1['EMA_0.1'] = df1['average'].ewm(alpha=0.1, adjust=False).mean()
df1.head()
```

```python
df1['EMA_0.3'] = df1['average'].ewm(alpha=0.3, adjust=False).mean()
df1.head()
```

```python
df1['EMA_S_0.1'] = df1['Sum'].ewm(alpha=0.1, adjust=False).mean()
df1.head()
```

```python
df1['EMA_S_0.3'] = df1['Sum'].ewm(alpha=0.3, adjust=False).mean()
df1.head()
```

```python
colors = ['green', 'orchid', 'orange']

# line plot - the yearly average air temperature in Barcelona
df1[['average', 'EMA_0.1', 'EMA_0.3']].plot(color=colors, linewidth=3, figsize=(12,6), alpha=0.8)

# modify ticks size
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.legend(labels=['Average', 'EMA - alpha=0.1', 'EMA - alpha=0.3'], fontsize=14)

# title and labels
plt.title('IEX', fontsize=20)
plt.xlabel('Entries', fontsize=16)
plt.ylabel('average', fontsize=16)
```

    - As the alpha values changes the exponetial graph also changes the more the value the more is the weight of the graph
